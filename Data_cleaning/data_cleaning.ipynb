```
import re
import nltk
from nltk.corpus import stopwords
import string
from textblob import TextBlob
#from autocorrect import spell
from nltk.stem import PorterStemmer
porterStemmer = PorterStemmer()
def data_cleaning(dataframe,coulmn):
    """
    this method cleaning text data :
    -remove punctuaion  
    -convert to Lower case
    - remove stope wordes 
    - make stemming to each word
    Parameter :
    -Data Frame 
    -Column Name in data frame
    Return :
    New data frame with cleaned column

    """

    dataframe[coulmn]= dataframe[coulmn].apply(lambda word : re.sub(r'[0-9]','',re.sub(r'[^\w\s]','',word)))
    dataframe[coulmn]=dataframe[coulmn].apply(lambda text: " ".join(word.lower() for word in text.split())) 
    stop = stopwords.words('english')
    dataframe[coulmn]=dataframe[coulmn].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
    dataframe[coulmn]=dataframe[coulmn].apply(lambda x: " ".join([porterStemmer.stem(word) for word in x.split()]))
    print(dataframe[coulmn])
    return dataframe
coulmn_cleaned = data_cleaning(df,'Course_Description')
```
