```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',markers=['o','v'])
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',markers=['o','v'],scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='time',row='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='autumn')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='autumn',aspect=.8,height=20)
```
